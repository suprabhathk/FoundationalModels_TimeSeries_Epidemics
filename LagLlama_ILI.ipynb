```python
import sys
from types import ModuleType

# Create dummy module hierarchy
def create_dummy_module(module_path):
    """
    Create a dummy module hierarchy for the given path.
    Returns the leaf module.
    """
    parts = module_path.split('.')
    current = ''
    parent = None

    for part in parts:
        current = current + '.' + part if current else part
        if current not in sys.modules:
            module = ModuleType(current)
            sys.modules[current] = module
            if parent:
                setattr(sys.modules[parent], part, module)
        parent = current

    return sys.modules[module_path]

# Create the dummy gluonts module hierarchy
gluonts_module = create_dummy_module('gluonts.torch.modules.loss')

# Create dummy classes for the specific loss functions
class DistributionLoss:
    def __init__(self, *args, **kwargs):
        pass

    def __call__(self, *args, **kwargs):
        return 0.0

    def __getattr__(self, name):
        return lambda *args, **kwargs: None

class NegativeLogLikelihood:
    def __init__(self, *args, **kwargs):
        pass

    def __call__(self, *args, **kwargs):
        return 0.0

    def __getattr__(self, name):
        return lambda *args, **kwargs: None

# Add the specific classes to the module
gluonts_module.DistributionLoss = DistributionLoss
gluonts_module.NegativeLogLikelihood = NegativeLogLikelihood
```

### 3. Lag Llama Prediction Function

```python
def get_lag_llama_predictions(dataset, prediction_length, device, context_length=256, use_rope_scaling=True, num_samples=100): # Changed context_length= 64 instead of 32

    # Add the StudentTOutput class to safe globals
    add_safe_globals([StudentTOutput])

    # Explicitly add NegativeLogLikelihood to safe globals
    add_safe_globals([NegativeLogLikelihood])

    # Load the checkpoint with weights_only=False
    ckpt = torch.load("lag-llama.ckpt", map_location=device, weights_only=False)  # Changed to False
    estimator_args = ckpt["hyper_parameters"]["model_kwargs"]

    rope_scaling_arguments = {
        "type": "linear",
        "factor": max(1.0, (context_length + prediction_length) / estimator_args["context_length"]),
    }

    estimator = LagLlamaEstimator(
        ckpt_path="lag-llama.ckpt",
        prediction_length=prediction_length,
        context_length=context_length, # Lag-Llama was trained with a context length of 32, but can work with any context length

        # estimator args
        input_size=estimator_args["input_size"],
        n_layer=estimator_args["n_layer"],
        n_embd_per_head=estimator_args["n_embd_per_head"],
        n_head=estimator_args["n_head"],
        scaling=estimator_args["scaling"],
        time_feat=estimator_args["time_feat"],
        rope_scaling=rope_scaling_arguments if use_rope_scaling else None,

        batch_size=1,
        num_parallel_samples=100,
        device=device,
    )

    lightning_module = estimator.create_lightning_module()
    transformation = estimator.create_transformation()
    predictor = estimator.create_predictor(transformation, lightning_module)

    forecast_it, ts_it = make_evaluation_predictions(
        dataset=dataset,
        predictor=predictor,
        num_samples=num_samples
    )
    forecasts = list(forecast_it)
    tss = list(ts_it)

    return forecasts, tss
```

### 4. Cleaning the raw data

```python
import pandas as pd
import numpy as np
from datetime import datetime
from gluonts.dataset.common import ListDataset
import torch
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from tqdm import tqdm
```

```python
def prepare_scenario_data(ili_data, input_scenario, scenario_name):
    """
    Prepare data for a specific scenario.
    """
    # Get scenario details
    scenario = input_scenario[input_scenario['Scenario'] == scenario_name].iloc[0]
    dstart_test = pd.to_datetime(scenario['dstart_test'])
    dend_test = pd.to_datetime(scenario['dend_test'])

    # Convert timestamp to datetime if it's not already
    ili_data['timestamp'] = pd.to_datetime(ili_data['timestamp'])

    # Calculate prediction length (in weeks)
    prediction_length = (dend_test - dstart_test).days // 7

    # Create GluonTS ListDataset for training
    test_data = ListDataset(
        [{"target": ili_data[ili_data['timestamp'] <= dend_test]['target'].values,
          "start": pd.Timestamp(ili_data['timestamp'].iloc[0]),
          "item_id": "ILI"}],
        freq="W"
    )

    # Extract ground truth for prediction period
    ground_truth = ili_data[
        (ili_data['timestamp'] > dstart_test) &
        (ili_data['timestamp'] <= dend_test)
    ]['target'].values

    # Create date range for predictions
    prediction_dates = pd.date_range(start=dstart_test, end=dend_test, freq='W')[1:]

    return test_data, prediction_length, ground_truth, prediction_dates
```

```python
def create_prediction_dataframe(forecasts, ground_truth, prediction_dates, scenario_name):
    """
    Create a DataFrame with predictions and confidence bounds.
    """
    # Get predictions and confidence bounds
    predictions = forecasts[0].samples

    # Ensure all arrays have the same length
    n = min(len(ground_truth), len(prediction_dates), predictions.shape[1])

    mean_predictions = np.mean(predictions, axis=0)[:n]
    lower_bound = np.percentile(predictions, 10, axis=0)[:n]
    upper_bound = np.percentile(predictions, 90, axis=0)[:n]

    # Create DataFrame with truncated arrays
    df = pd.DataFrame({
        'Data': ['ILI'] * n,
        'Scenario': [scenario_name] * n,
        'date': prediction_dates[:n],
        'inc': ground_truth[:n],
        'y': mean_predictions,
        'y_lo': lower_bound,
        'y_hi': upper_bound
    })

    return df
```

```python
def generate_all_scenarios_predictions(ili_data, input_scenarios):
    """
    Generate predictions for all scenarios and combine into one DataFrame.

    Args:
        ili_data: Complete ILI dataset
        input_scenarios: DataFrame containing scenario information

    Returns:
        Combined DataFrame with predictions for all scenarios
    """
    # Initialize empty list to store DataFrames
    all_predictions = []

    # Get device
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

    # Loop through each scenario
    for idx, row in tqdm(input_scenarios.iterrows(), total=len(input_scenarios), desc="Processing scenarios"):
        scenario_name = row['Scenario']

        try:
            # Prepare data for this scenario
            test_data, prediction_length, ground_truth, prediction_dates = prepare_scenario_data(
                ili_data,
                input_scenarios,
                scenario_name
            )

            # Make predictions
            forecasts, tss = get_lag_llama_predictions(
                dataset=test_data,
                prediction_length=prediction_length,
                device=device
            )

            # Create DataFrame for this scenario
            scenario_df = create_prediction_dataframe(
                forecasts,
                ground_truth,
                prediction_dates,
                scenario_name
            )

            # Append to list
            all_predictions.append(scenario_df)

        except Exception as e:
            print(f"Error processing scenario {scenario_name}: {str(e)}")
            continue

    # Combine all predictions
    final_df = pd.concat(all_predictions, ignore_index=True)

    return final_df
```

```python
# Get predictions for all scenarios
ili_scenarios = input_ili[input_ili['Data'] == 'ILI']
all_predictions_df = generate_all_scenarios_predictions(ili_lagllama, ili_scenarios)
```
